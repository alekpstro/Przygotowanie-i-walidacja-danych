```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.stats import chi2_contingency, f_oneway

RAW_PATH = Path("adultData.csv")
CLEAN_PATH = Path("adultDataClean.csv")


# =============================================================================
# Ładowanie i bazowe czyszczenie
# =============================================================================

def load_raw(path: Path = RAW_PATH) -> pd.DataFrame:
    return pd.read_csv(path)


def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out.columns = out.columns.str.strip()
    return out


def drop_unused(df: pd.DataFrame) -> pd.DataFrame:
    return df.drop(columns=["fnlwgt"], errors="ignore")


def replace_missing_and_strip(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out.replace([" ?", "?"], np.nan, inplace=True)
    obj_cols = out.select_dtypes(include="object").columns
    out[obj_cols] = out[obj_cols].apply(lambda s: s.str.strip())
    return out


def unify_income_column(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    alt_targets = [c for c in ["salary", "class"] if c in out.columns]
    if "income" in out.columns:
        out = out.drop(columns=alt_targets, errors="ignore")
    elif alt_targets:
        primary = alt_targets[0]
        out = out.rename(columns={primary: "income"})
        extras = [c for c in alt_targets[1:] if c in out.columns]
        out = out.drop(columns=extras, errors="ignore")
    return out


def clean_adult(path: Path = RAW_PATH) -> pd.DataFrame:
    df = load_raw(path)
    df = normalize_columns(df)
    df = drop_unused(df)
    df = replace_missing_and_strip(df)
    df = unify_income_column(df)
    return df


adult_base = clean_adult()
print("Bazowe kolumny:", adult_base.columns.tolist())
print("Braki danych (bazowe):\n", adult_base.isna().sum())
print("Podgląd bazowy:")
display(adult_base.head())


# =============================================================================
# Profilowanie zbioru
# =============================================================================

def opis_zbioru(df: pd.DataFrame, nazwa: str = "adult") -> None:
    n_rows, n_cols = df.shape
    total_missing = int(df.isna().sum().sum())
    missing_pct = (total_missing / (n_rows * n_cols) * 100) if (n_rows * n_cols) else 0.0
    print(f"Opis zbioru danych: {nazwa}")
    print(f"- Wiersze: {n_rows}")
    print(f"- Kolumny: {n_cols}")
    print(f"- Braki łącznie: {total_missing} ({missing_pct:.2f}%)")
    print("- Typy danych:")
    print(df.dtypes.value_counts())


def typ_analityczny(series: pd.Series) -> str:
    return "ilościowa" if pd.api.types.is_numeric_dtype(series) else "jakościowa"


def skala_pomiarowa(df: pd.DataFrame, col: str) -> str:
    ordinal = {"education": "porządkowa"}
    ratio = {"age", "capital-gain", "capital-loss", "hours-per-week", "education-num"}
    nominal = {"workclass", "occupation", "native-country", "marital-status", "relationship", "race", "sex", "income", "salary", "class"}
    if col in ordinal:
        return ordinal[col]
    if col in ratio:
        return "ilorazowa"
    if col in nominal:
        return "nominalna"
    return "ilorazowa" if pd.api.types.is_numeric_dtype(df[col]) else "nominalna"


def charakterystyka_zmiennych(df: pd.DataFrame, sample_n: int = 5) -> pd.DataFrame:
    rows = []
    n = len(df)
    for col in df.columns:
        s = df[col]
        missing = int(s.isna().sum())
        missing_pct = (missing / n * 100) if n else 0.0
        examples = s.dropna().astype(str).unique()[:sample_n].tolist()
        rows.append({
            "zmienna": col,
            "typ_pandas": str(s.dtype),
            "typ_analityczny": typ_analityczny(s),
            "skala_pomiarowa": skala_pomiarowa(df, col),
            "unikaty_bez_NaN": int(s.nunique(dropna=True)),
            "unikaty_z_NaN": int(s.nunique(dropna=False)),
            "braki_NaN": missing,
            "braki_%": round(missing_pct, 2),
            "przykładowe_wartości": ", ".join(examples),
        })
    meta = pd.DataFrame(rows).sort_values(["typ_analityczny", "zmienna"]).reset_index(drop=True)
    return meta

opis_zbioru(adult_base, nazwa="adultData.csv (clean bazowy)")
meta = charakterystyka_zmiennych(adult_base, sample_n=6)
display(meta)


# =============================================================================
# Zależności: Cramér's V i ANOVA
# =============================================================================

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2_corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    r_corr = r - ((r - 1) ** 2) / (n - 1)
    k_corr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2_corr / min(k_corr - 1, r_corr - 1))

categorical1_vars = ["workclass", "sex", "race", "education"]
print("Cramér's V: occupation vs ...")
for col in categorical1_vars:
    mask = adult_base["occupation"].notna()
    v = cramers_v(adult_base.loc[mask, "occupation"], adult_base.loc[mask, col])
    print(f"occupation vs {col}: {v:.3f}")

categorical2_vars = ["occupation", "sex", "race", "education"]
print("\nCramér's V: workclass vs ...")
for col in categorical2_vars:
    mask = adult_base["workclass"].notna()
    v = cramers_v(adult_base.loc[mask, "workclass"], adult_base.loc[mask, col])
    print(f"workclass vs {col}: {v:.3f}")

numerical_cols = adult_base.select_dtypes(include=np.number).columns
print("\nANOVA: occupation vs numeryczne")
for col in numerical_cols:
    groups = [adult_base.loc[adult_base["occupation"] == occ, col] for occ in adult_base["occupation"].dropna().unique()]
    f_stat, p_val = f_oneway(*groups)
    print(f"occupation vs {col}: F={f_stat:.2f}, p={p_val:.2e}")

print("\nANOVA: workclass vs numeryczne")
for col in numerical_cols:
    groups = [adult_base.loc[adult_base["workclass"] == wc, col] for wc in adult_base["workclass"].dropna().unique()]
    f_stat, p_val = f_oneway(*groups)
    print(f"workclass vs {col}: F={f_stat:.2f}, p={p_val:.2e}")


# =============================================================================
# Imputacje i transformacje
# =============================================================================

def hot_deck_impute(df: pd.DataFrame, target: str, cell_levels: list[list[str]], seed: int = 42) -> pd.DataFrame:
    rng = np.random.default_rng(seed)
    out = df.copy()
    missing_mask = out[target].isna()
    donors = out.loc[~missing_mask]
    for idx in out.loc[missing_mask].index:
        for cols in cell_levels:
            pool = donors if len(cols) == 0 else donors.loc[(donors[cols] == out.loc[idx, cols]).all(axis=1)]
            if len(pool) > 0:
                donor = pool.sample(1, random_state=int(rng.integers(0, 1_000_000)))
                out.at[idx, target] = donor[target].values[0]
                break
    return out


def fill_with_group_mode(df: pd.DataFrame, target: str, by: list[str]) -> pd.DataFrame:
    out = df.copy()
    miss = out[target].isna()
    non_na = out[target].dropna()
    fallback_mode = non_na.mode().iloc[0] if len(non_na) else np.nan

    def _mode_or_nan(x: pd.Series):
        x = x.dropna()
        if len(x) == 0:
            return np.nan
        m = x.mode()
        return m.iloc[0] if len(m) else np.nan

    group_mode = out.groupby(by, dropna=False)[target].transform(_mode_or_nan)
    out.loc[miss, target] = group_mode[miss]
    out[target] = out[target].fillna(fallback_mode)
    return out


def remove_outliers_tukey(df: pd.DataFrame, columns: list[str]) -> pd.DataFrame:
    out = df.copy()
    n_initial = len(out)
    for col in columns:
        n_before = len(out)
        Q1 = out[col].quantile(0.25)
        Q3 = out[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        out = out[(out[col] >= lower) & (out[col] <= upper)]
        print(f"{col}: usunięto {n_before - len(out)} wierszy (przedział [{lower:.2f}, {upper:.2f}])")
    print(f"Outliery: start={n_initial}, koniec={len(out)}, łączna różnica={n_initial - len(out)}")
    return out


def iqr_bounds(series: pd.Series) -> tuple[float, float]:
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower, upper


def detect_outliers_iqr(df: pd.DataFrame, columns: list[str]) -> pd.DataFrame:
    rows = []
    n = len(df)
    for col in columns:
        lower, upper = iqr_bounds(df[col])
        mask = (df[col] < lower) | (df[col] > upper)
        count = int(mask.sum())
        pct = round(count / n * 100, 2) if n else 0.0
        rows.append({"kolumna": col, "metoda": "IQR", "dolna": lower, "górna": upper, "liczba": count, "udział_%": pct})
    return pd.DataFrame(rows)


def detect_outliers_zscore(df: pd.DataFrame, columns: list[str], threshold: float = 3.0) -> pd.DataFrame:
    rows = []
    n = len(df)
    for col in columns:
        mean = df[col].mean()
        std = df[col].std()
        if std == 0 or np.isnan(std):
            count = 0
            bounds = (np.nan, np.nan)
        else:
            z = (df[col] - mean) / std
            mask = z.abs() > threshold
            count = int(mask.sum())
            bounds = (mean - threshold * std, mean + threshold * std)
        pct = round(count / n * 100, 2) if n else 0.0
        rows.append({"kolumna": col, "metoda": f"Z-score>|{threshold}|", "dolna": bounds[0], "górna": bounds[1], "liczba": count, "udział_%": pct})
    return pd.DataFrame(rows)


adult_imputed = adult_base.copy()

# Hot Deck: occupation
adult_imputed = hot_deck_impute(
    adult_imputed,
    target="occupation",
    cell_levels=[["sex", "workclass", "education"], ["sex", "workclass"], ["education"], []],
    seed=42,
)

# Hot Deck: workclass
adult_imputed = hot_deck_impute(
    adult_imputed,
    target="workclass",
    cell_levels=[["sex", "occupation", "education"], ["sex", "education"], ["sex"], ["education"], []],
    seed=123,
)

# Moda grupowa: native-country wg race
adult_imputed = fill_with_group_mode(adult_imputed, target="native-country", by=["race"])

# Grupowanie wieku zgodnie z żądanymi przedziałami
age_bins = [15, 29, 59, np.inf]
age_labels = ["Młodzi (15–29)", "Dorośli (30–59)", "Starsi (60+)"]
adult_imputed["age_category"] = pd.cut(
    adult_imputed["age"],
    bins=age_bins,
    labels=age_labels,
    right=True,
    include_lowest=True,
)

# Wizualizacja: struktura wiekowa vs płeć
adult_plot = adult_imputed[["sex", "age_category"]].dropna().copy()
ct = pd.crosstab(adult_plot["sex"], adult_plot["age_category"]).reindex(columns=age_labels, fill_value=0)
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(1, 2, figsize=(12, 4), gridspec_kw={"width_ratios": [1, 1.2]})
ct.plot(kind="bar", stacked=True, ax=axes[0], colormap="tab20")
axes[0].set_title("Liczebność grup wiekowych w podziale na płeć")
axes[0].set_xlabel("Płeć")
axes[0].set_ylabel("Liczba osób")
axes[0].grid(axis="y", alpha=0.2)
axes[0].legend(title="Grupa wiekowa", bbox_to_anchor=(1.02, 1), loc="upper left")

ct_pct.plot(kind="bar", stacked=True, ax=axes[1], colormap="tab20")
axes[1].set_title("Struktura wiekowa w obrębie płci (udział %)")
axes[1].set_xlabel("Płeć")
axes[1].set_ylabel("Udział (%)")
axes[1].set_ylim(0, 100)
axes[1].grid(axis="y", alpha=0.2)
axes[1].legend(title="Grupa wiekowa", bbox_to_anchor=(1.02, 1), loc="upper left")

plt.tight_layout()
plt.show()

# Binowanie sex na 0/1 po wizualizacji
adult_imputed["sex"] = (adult_imputed["sex"] == "Male").astype(int)

# Analiza wartości nietypowych (bez usuwania) dla wybranych kolumn
outlier_cols = ["age", "hours-per-week", "capital-gain", "capital-loss"]
summary_iqr = detect_outliers_iqr(adult_imputed, outlier_cols)
summary_z = detect_outliers_zscore(adult_imputed, outlier_cols, threshold=3.0)
outlier_summary = pd.concat([summary_iqr, summary_z], ignore_index=True)
print("Podsumowanie outlierów (IQR i Z-score):")
display(outlier_summary)

# Boxploty porównawcze
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, col in zip(axes.flat, outlier_cols):
    ax.boxplot(adult_imputed[col].dropna(), vert=True, patch_artist=True)
    ax.set_title(f"Boxplot: {col}")
    ax.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

adult_processed = adult_imputed.copy()

print("Braki po imputacji (bez usuwania outlierów):\n", adult_processed.isna().sum())
print("Podgląd po imputacji:")
display(adult_processed.head())


# =============================================================================
# Zapis do pliku (przed transformacjami capital-gain/loss)
# =============================================================================

adult_processed.to_csv(CLEAN_PATH, index=False)
print(f"Zapisano oczyszczony zbiór (bez transformacji capital) do: {CLEAN_PATH.resolve()}")
```

```python
##### Z analizy wyszło że capital-gain i capital-loss mają rozkład pełny zero przez co warto je przekształcić aby układ był mniej skośny.


# Transformacje capital-gain i capital-loss: log(x+1) aby spłaszczyć rozkład zero-inflated
adult_imputed["capital-gain_log"] = np.log1p(adult_imputed["capital-gain"])
adult_imputed["capital-loss_log"] = np.log1p(adult_imputed["capital-loss"])

# Wizualizacja przed/po transformacji
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Capital-gain przed
axes[0, 0].hist(adult_imputed["capital-gain"].dropna(), bins=50, edgecolor="black", alpha=0.7)
axes[0, 0].set_title("capital-gain (oryginał)")
axes[0, 0].grid(axis="y", alpha=0.3)

# Capital-gain po
axes[0, 1].hist(adult_imputed["capital-gain_log"].dropna(), bins=50, edgecolor="black", alpha=0.7, color="orange")
axes[0, 1].set_title("capital-gain (log transformacja)")
axes[0, 1].grid(axis="y", alpha=0.3)

# Capital-loss przed
axes[1, 0].hist(adult_imputed["capital-loss"].dropna(), bins=50, edgecolor="black", alpha=0.7, color="red")
axes[1, 0].set_title("capital-loss (oryginał)")
axes[1, 0].grid(axis="y", alpha=0.3)

# Capital-loss po
axes[1, 1].hist(adult_imputed["capital-loss_log"].dropna(), bins=50, edgecolor="black", alpha=0.7, color="darkred")
axes[1, 1].set_title("capital-loss (log transformacja)")
axes[1, 1].grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()

# Statystyki przed/po
print("Statystyki capital-gain:")
print(f"Przed: średnia={adult_imputed['capital-gain'].mean():.2f}, std={adult_imputed['capital-gain'].std():.2f}, skośność={adult_imputed['capital-gain'].skew():.2f}")
print(f"Po: średnia={adult_imputed['capital-gain_log'].mean():.2f}, std={adult_imputed['capital-gain_log'].std():.2f}, skośność={adult_imputed['capital-gain_log'].skew():.2f}")

print("\nStatystyki capital-loss:")
print(f"Przed: średnia={adult_imputed['capital-loss'].mean():.2f}, std={adult_imputed['capital-loss'].std():.2f}, skośność={adult_imputed['capital-loss'].skew():.2f}")
print(f"Po: średnia={adult_imputed['capital-loss_log'].mean():.2f}, std={adult_imputed['capital-loss_log'].std():.2f}, skośność={adult_imputed['capital-loss_log'].skew():.2f}")

# Finalizacja zbioru do zapisania
adult_final = adult_imputed.copy()
```

```python
# =============================================================================
# Podstawowe charakterystyki całego oczyszczonego zbioru
# =============================================================================

print("\n" + "="*80)
print("PODSTAWOWE CHARAKTERYSTYKI ZBIORU DANYCH (adultDataClean)")
print("="*80)

# 1. Wymiary zbioru
print(f"\n1. WYMIARY ZBIORU:")
print(f"   Liczba wierszy: {len(adult_final):,}")
print(f"   Liczba kolumn: {len(adult_final.columns)}")
print(f"   Braki danych (razem): {int(adult_final.isna().sum().sum())} ({adult_final.isna().sum().sum() / (len(adult_final) * len(adult_final.columns)) * 100:.2f}%)")

# 2. Statystyki zmiennych numerycznych
print(f"\n2. STATYSTYKI ZMIENNYCH NUMERYCZNYCH:")
numeric_stats = adult_final.describe().T
numeric_stats["skośność"] = adult_final.select_dtypes(include=np.number).skew()
numeric_stats["kurtoza"] = adult_final.select_dtypes(include=np.number).kurtosis()
display(numeric_stats)

# 3. Rozkład zmiennych kategorycznych (top 5 wartości)
print(f"\n3. ROZKŁAD ZMIENNYCH KATEGORYCZNYCH (top 5 wartości):")
categorical_cols = adult_final.select_dtypes(include="object").columns
for col in categorical_cols:
    print(f"\n   {col}:")
    freq = adult_final[col].value_counts().head(5)
    for val, count in freq.items():
        pct = count / len(adult_final) * 100
        print(f"      {val}: {count:,} ({pct:.2f}%)")

# 4. Braki danych na kolumnę
print(f"\n4. BRAKI DANYCH NA KOLUMNĘ:")
missing = adult_final.isna().sum()
missing_pct = missing / len(adult_final) * 100
missing_df = pd.DataFrame({
    "kolumna": missing.index,
    "liczba_braków": missing.values,
    "udział_%": missing_pct.values
}).sort_values("liczba_braków", ascending=False)
missing_df = missing_df[missing_df["liczba_braków"] > 0]
if len(missing_df) > 0:
    display(missing_df)
else:
    print("   Brak braków danych w zbiorze!")

# 5. Korelacje zmiennych numerycznych
print(f"\n5. MACIERZ KORELACJI (zmienne numeryczne):")
numeric_cols = adult_final.select_dtypes(include=np.number).columns
corr_matrix = adult_final[numeric_cols].corr()
display(corr_matrix)

# 6. Top korelacje (bez przekątnej)
print(f"\n6. TOP 10 NAJBARDZIEJ SKORELOWANYCH PAR ZMIENNYCH:")
corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        corr_pairs.append({
            "zmienna_1": corr_matrix.columns[i],
            "zmienna_2": corr_matrix.columns[j],
            "korelacja": corr_matrix.iloc[i, j]
        })
corr_pairs_df = pd.DataFrame(corr_pairs).sort_values("korelacja", key=abs, ascending=False)
display(corr_pairs_df.head(10))

# 7. Rozkład zmiennej celu (income)
if "income" in adult_final.columns:
    print(f"\n7. ROZKŁAD ZMIENNEJ CELU (income):")
    income_dist = adult_final["income"].value_counts()
    income_pct = adult_final["income"].value_counts(normalize=True) * 100
    for val in income_dist.index:
        print(f"   {val}: {income_dist[val]:,} ({income_pct[val]:.2f}%)")

# 8. Wizualizacja rozkładów zmiennych numerycznych
print(f"\n8. WYKRESY ROZKŁADÓW ZMIENNYCH NUMERYCZNYCH:")
numeric_cols_plot = adult_final.select_dtypes(include=np.number).columns
n_cols = min(len(numeric_cols_plot), 5)
fig, axes = plt.subplots(
    (len(numeric_cols_plot) + n_cols - 1) // n_cols,
    n_cols,
    figsize=(15, 3 * ((len(numeric_cols_plot) + n_cols - 1) // n_cols))
)
axes = axes.flatten()
for i, col in enumerate(numeric_cols_plot):
    axes[i].hist(adult_final[col].dropna(), bins=50, edgecolor="black", alpha=0.7, color="steelblue")
    axes[i].set_title(f"Rozkład: {col}")
    axes[i].grid(axis="y", alpha=0.3)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Częstość")

for j in range(i+1, len(axes)):
    axes[j].axis("off")

plt.tight_layout()
plt.show()

# 9. Wizualizacja rozkładów zmiennych kategorycznych
print(f"\n9. WYKRESY ROZKŁADÓW ZMIENNYCH KATEGORYCZNYCH:")
categorical_cols_plot = adult_final.select_dtypes(include="object").columns
n_cols = min(len(categorical_cols_plot), 4)
fig, axes = plt.subplots(
    (len(categorical_cols_plot) + n_cols - 1) // n_cols,
    n_cols,
    figsize=(16, 4 * ((len(categorical_cols_plot) + n_cols - 1) // n_cols))
)
axes = axes.flatten()
for i, col in enumerate(categorical_cols_plot):
    top_vals = adult_final[col].value_counts().head(10)
    axes[i].barh(range(len(top_vals)), top_vals.values, color="coral")
    axes[i].set_yticks(range(len(top_vals)))
    axes[i].set_yticklabels(top_vals.index)
    axes[i].set_title(f"Top 10: {col}")
    axes[i].grid(axis="x", alpha=0.3)

for j in range(i+1, len(axes)):
    axes[j].axis("off")

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("KONIEC CHARAKTERYSTYKI")
print("="*80)
```
